```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3)
qc.ccx(0, 1, 2)
qc.cx( 1,2)
qc.x(2)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(12)

# Map the qubits to their corresponding labels in the diagram
# a: ancilla qubits
# p: primary qubits
# s: sum qubits
# c: carry qubits

# Apply the controlled-X gates as shown in the diagram
qc.cx(0, 3)  # Controlled-X with control on a_0 and target on p_0
qc.cx(1, 4)  # Controlled-X with control on a_1 and target on p_1
qc.cx(2, 5) 

qc.draw(output='mpl')
```
